```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
import glob



data_aug_gen = ImageDataGenerator(rotation_range=360, # 0~ 360도사이 임의의 각도로 회전
                                  width_shift_range=0.1, # 원본이미지 0.1픽셀 내외로 좌우이동
                                  height_shift_range=0.1, # 원본이미지 0.1픽셀 내외로 상하이동
                                  shear_range=0.1, # X라디안 시계반대방향으로 밀림강도 범위내에서 원본이미지 변경
                                  zoom_range= 0.2, # 확대(1+X)/축소(1-X)
                                  horizontal_flip=True,#수평방향 뒤집기
                                  vertical_flip=True,#수직방향 뒤집기
                                  fill_mode='nearest')

file=glob.glob("C:/Users/user/PycharmProjects/donghwi/donghwikim/code/OCT2017/train/make_img/" + "*")

for l ,f in enumerate(file):
    img = load_img(f)
    x = img_to_array(img)
    x = x.reshape((1,) + x.shape)
    print(f + "변형된 이미지 생성")
    count=0
    for batch in data_aug_gen.flow(x, save_to_dir='D:/YoonHwan/make_img/', save_prefix='remake_1', save_format='jpeg'):
        count += 1
        if count > 1:
            break
```
