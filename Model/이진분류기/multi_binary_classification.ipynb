# Normal_Binary 

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten, Dropout, BatchNormalization
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.optimizers import Adam
from keras.preprocessing.image import ImageDataGenerator

def first_model():
    model = Sequential()
    model.add(Conv2D(32, kernel_size=(3, 3), padding='same', input_shape=(299, 299, 1), activation='relu'))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.25))
    model.add(Conv2D(64, kernel_size=(3, 3), padding="same", activation='relu'))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.25))
    model.add(Flatten())
    model.add(Dense(256, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.25))
    model.add(Dense(1, activation='sigmoid'))
    return model

train_datagen = ImageDataGenerator(rescale=1./255)
train_generator = train_datagen.flow_from_directory(
        'D:/OCT Project/OCT-Image/이진분류기 이미지/NORMAL_BINARY/train',
        target_size=(299, 299),
        color_mode = 'grayscale',
        batch_size=64,                                
        class_mode='binary')

model1 = first_model()
model1.compile(loss='binary_crossentropy',
                         optimizer=Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0),
                         metrics=['acc'])

model1.summary()

print("start train")
model1.fit_generator(train_generator, epochs = 100)
print("train done")

model1.save("D:/OCT Project/OCT-Image/이진분류기 이미지/normal_binary.h5")
```

```python
import numpy as np
from PIL import Image
import glob

def image_processing(img, color, w, h, chanel):
    img = Image.open(img)
    img = img.convert(color)
    img = img.resize((w,h))
    data = np.asarray(img)
    data = data.reshape((1,w,h,chanel))
    data = data / 255
    return data

test_path = "D:/OCT Project/OCT-Image/total1/test/"
categories = ['CNV', 'DME', 'DRUSEN', 'NORMAL']

from keras.models import load_model
model_path = 'D:/OCT Project/OCT-Image/이진분류기 이미지/'
model1 = load_model(model_path + 'normal_binary.h5')
model2 = load_model(model_path + 'CNV_binary.h5')
model3 = load_model(model_path + 'DME_binary.h5')


for idx, categorie in enumerate(categories):
    
    total = []
    cnv = []
    dme = []
    drusen = []
    normal = []
    
    files = glob.glob(test_path + categorie + "/*") 
    for idx2, file in enumerate(files):
        data = image_processing(file, 'L', 299, 299, 1)
        result = model1.predict_classes(data)

        if result[0] == 0:
            normal.append(result[0])
        if result[0] == 0 and categorie == 'NORMAL':
            total.append(1)
        else:
            result2 = model2.predict_classes(data)
            if result2[0] == 0:
                cnv.append(result2[0])
            if result2[0] == 0 and categorie == 'CNV':
                total.append(1)
            else:
                result3 = model3.predict_classes(data)
                if result3[0] == 0:
                    total.append(1)
                if result3[0] == 0 and categorie == 'DME':
                    dme.append(result3[0])
                else:
                    if result3[0] == 1:
                        drusen.append(result3[0])
                    if result3[0] == 1 and categorie == 'DRUSEN':
                        total.append(1)
                    else:
                        total.append(0)
    
    
    print(total)
    print("평균 : " + str(np.mean(total)))
```
